```python
#Importar librerías
import pandas as pd
import random
from datetime import datetime
```

```python
product_catalog = pd.read_csv("Chocolate_Data.csv")#, parse_dates=["Date"])
product_catalog
```

```python
holiday_sales_boost = [
    {
        "Holiday": "Valentine's Day",
        "DateRange": ("2023-02-07", "2023-02-14"),
        "SalesIncreasePercent": 40
    },
    {
        "Holiday": "Easter",
        "DateRange": ("2023-04-02", "2023-04-09"),
        "SalesIncreasePercent": 30
    },
    {
        "Holiday": "Halloween",
        "DateRange": ("2023-10-24", "2023-10-31"),
        "SalesIncreasePercent": 25
    },
    {
        "Holiday": "Christmas",
        "DateRange": ("2023-12-15", "2023-12-25"),
        "SalesIncreasePercent": 35
    }
]
```

```python
# Rango de fechas
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 12, 31)
date_range = pd.date_range(start=start_date, end=end_date)

# Valores únicos
products = product_catalog['ProductID'].tolist()
sales_channels = ['Online', 'Retail', 'Wholesale']
promotion_types = ['None', 'Bundle']

# Crear combinaciones y generar datos
sales_data = []

for date in date_range:
    for product_id in products:
        for channel in sales_channels:
            for promo in promotion_types:
                units_sold = random.randint(0, 50)
                sales_data.append([
                    date.date(), product_id, channel, promo, units_sold
                ])

# Crear DataFrame
sales_df = pd.DataFrame(sales_data, columns=[
    "Date", "ProductID", "SalesChannel", "PromotionType", "UnitsSold"
])

# Mostrar parte del resultado
print(sales_df.head())
```

```python
# Asegúrate de que sales_df tenga la columna 'Date' como datetime
sales_df['Date'] = pd.to_datetime(sales_df['Date'])

# Lista de festividades con rango de fechas y aumento de ventas en %
holiday_sales_boost = [
    {
        "Holiday": "Valentine's Day",
        "DateRange": ("2023-02-07", "2023-02-14"),
        "SalesIncreasePercent": 40
    },
    {
        "Holiday": "Easter",
        "DateRange": ("2023-04-02", "2023-04-09"),
        "SalesIncreasePercent": 30
    },
    {
        "Holiday": "Halloween",
        "DateRange": ("2023-10-24", "2023-10-31"),
        "SalesIncreasePercent": 25
    },
    {
        "Holiday": "Christmas",
        "DateRange": ("2023-12-15", "2023-12-25"),
        "SalesIncreasePercent": 35
    }
]

# Inicializar columna de festivos como None
sales_df['Holiday'] = None

# Aplicar el ajuste de ventas según la festividad
for event in holiday_sales_boost:
    start_date = pd.to_datetime(event["DateRange"][0])
    end_date = pd.to_datetime(event["DateRange"][1])
    percent_increase = event["SalesIncreasePercent"]

    mask = (sales_df["Date"] >= start_date) & (sales_df["Date"] <= end_date)
    sales_df.loc[mask, "UnitsSold"] = (sales_df.loc[mask, "UnitsSold"] * (1 + percent_increase / 100)).round().astype(int)
    sales_df.loc[mask, "Holiday"] = event["Holiday"]
```

```python
sales_df.to_csv("Chocolate_Sales_Dataset_2023.csv",index=False)
```
