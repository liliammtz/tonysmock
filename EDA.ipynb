```python
#Importar librer√≠as
import pandas as pd
import random
from datetime import datetime
```

```python
product_catalog = pd.read_csv("Chocolate_Data.csv")
sales = pd.read_csv("Chocolate_Sales_Dataset_2023.csv", parse_dates=["Date"])
```

```python
df = sales.merge(product_catalog, on='ProductID',how='left')

df.head(5)
```

```python
# Feature engineering for date column:
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["MonthName"] = df["Date"].dt.strftime('%B')
df["Day"] = df["Date"].dt.day
df["Week"] = df["Date"].dt.isocalendar().week
df["Quarter"] = df["Date"].dt.quarter
df["DayOfWeek"] = df["Date"].dt.dayofweek
df["WeekdayName"] = df["Date"].dt.strftime('%A')
```

```python
df.to_csv("Database.csv",index=False)
```

```python
weekly_sales = df.groupby('WeekdayName')['UnitsSold'].sum().reset_index().sort_values('UnitsSold', ascending=False)
weekly_sales['WeekdayName'][0]
weekly_sales['UnitsSold'][0]
```
